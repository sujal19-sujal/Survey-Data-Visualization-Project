```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load the dataset
survey_df = pd.read_csv('survey_data.csv')

# Display the first few rows and information about the DataFrame
print("Survey Data DataFrame:")
print(survey_df.head())
print("\nSurvey Data Info:")
survey_df.info()
```

```python
# Create a figure with two subplots for gender and feature preference
fig, axes = plt.subplots(1, 2, figsize=(16, 6))

# Plot for Gender distribution
sns.countplot(x='gender', data=survey_df, ax=axes[0], palette='pastel')
axes[0].set_title('Gender Distribution', fontsize=16)
axes[0].set_xlabel('Gender', fontsize=12)
axes[0].set_ylabel('Count', fontsize=12)
axes[0].tick_params(axis='x', rotation=45)

# Plot for Feature Preference
sns.countplot(x='feature_preference', data=survey_df, ax=axes[1], palette='viridis')
axes[1].set_title('Feature Preference', fontsize=16)
axes[1].set_xlabel('Feature', fontsize=12)
axes[1].set_ylabel('Count', fontsize=12)
axes[1].tick_params(axis='x', rotation=45)

plt.tight_layout()
plt.show()
```

```python
# Count the frequencies of each referral likelihood response
referral_counts = survey_df['referral_likelihood'].value_counts()

# Create a pie chart
plt.figure(figsize=(8, 8))
plt.pie(referral_counts, labels=referral_counts.index, autopct='%1.1f%%', startangle=140, colors=sns.color_palette('rocket'))
plt.title('Referral Likelihood', fontsize=16)
plt.axis('equal') # Equal aspect ratio ensures that pie is drawn as a circle.
plt.show()
```

```python
# Select numerical columns for correlation analysis
numerical_df = survey_df[['age', 'satisfaction_rating']]

# Calculate the correlation matrix
correlation_matrix = numerical_df.corr()

# Create a heatmap
plt.figure(figsize=(8, 6))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f", linewidths=.5)
plt.title('Correlation Heatmap: Age vs. Satisfaction Rating', fontsize=16)
plt.show()
```
